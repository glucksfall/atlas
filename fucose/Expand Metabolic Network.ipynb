```python
import pandas
```

```python
data = pandas.read_csv('./metabolic_network_v2.txt', header = None, delimiter = '\t')
data
```

```python
print('SOURCE	SOURCE_ATTRIBUTE	TARGET	TARGET_ATTRIBUTE')
for enzyme, reaction in zip(data.loc[:,0], data.loc[:,1]):
    print('{:s}\tGENE_PROD\t{:s}\tRXN'.format(enzyme, reaction))
print()
    
for reaction, substrates in zip(data.loc[:,1], data.loc[:,2]):
    for subtrate in substrates.split(', '):
        print('{:s}\tMET\t{:s}\tRXN'.format(subtrate, reaction))
print()
        
for reaction, products in zip(data.loc[:,1], data.loc[:,3]):
    for product in products.split(', '):
        print('{:s}\tRXN\t{:s}\tMET'.format(reaction, product))
```
